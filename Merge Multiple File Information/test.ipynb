```python
import pandas as pd
import numpy as np
import re
from datetime import datetime as dt
```

```python
fw_log = pd.read_csv(r"input_files\20230311_FW_LOG_1.log", sep=";")
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
fw_log.drop('Unnamed: 0', axis=1, inplace=True)

val1 = fw_log.columns
val2 = fw_log.iloc[0].to_list()
val3 = fw_log.iloc[1].to_list()

fw_log.drop([0, 1], axis=0, inplace=True)
new_header = [str(v1) + str(v2) + str(v3) for v1, v2, v3 in list(zip(val1, val2, val3))]
new_header = [re.sub(r"Unnamed: \d|YYYY/MM/DD|HH:MM:SS|nan", "", some_string) for some_string in new_header]
new_header = [re.sub(r" |\.", "", some_string) for some_string in new_header]
# new_header = [re.sub(r"t_acts|t_tots", "", some_string) for some_string in new_header]
# new_header = [re.sub(r"\.|\-", "_", some_string).lower() for some_string in new_header]
fw_log.columns = new_header

fw_log['Date'] = pd.to_datetime(fw_log["Date"] + " " + fw_log['Time'])
fw_log.drop('Time', axis=1, inplace=True)

print(len(fw_log))
print(fw_log.info())

# Resultados de operación.
fw_log.head()
```
